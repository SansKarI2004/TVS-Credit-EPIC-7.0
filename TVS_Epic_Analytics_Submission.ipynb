# Task
Perform exploratory data analysis (EDA) on the dataset located at "/content/Analytics_loan_collection_dataset.csv".

## Descriptive statistics

### Subtask:
Calculate and display descriptive statistics for numerical columns to understand the central tendency, dispersion, and shape of the data.


**Reasoning**:
Select numerical columns and calculate descriptive statistics.



```python
numerical_cols = df.select_dtypes(include=np.number)
display(numerical_cols.describe())
```

## Visualize distributions


Created visualizations (histograms, box plots) for numerical features to examine their distributions and identify potential outliers.


```python
for col in numerical_cols.columns:
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    plt.subplot(1, 2, 2)
    sns.boxplot(data=df, x=col)
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)

    plt.tight_layout()
    plt.show()
```

## Analyze categorical variables


Analyzed the distribution of categorical variables using frequency counts and bar plots.


```python
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"Frequency counts for {col}:")
    display(df[col].value_counts())
```

```python
for col in categorical_cols:
    if col != 'CustomerID':
        plt.figure(figsize=(10, 6))
        sns.countplot(data=df, x=col, order=df[col].value_counts().index)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
```

## Correlation analysis


Computed and visualized the correlation matrix to understand the relationships between numerical variables.


```python
correlation_matrix = numerical_cols.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## Target variable analysis


Analyzed the distribution of the target variable and explore its relationship with other features.


```python
target_counts = df['Target'].value_counts()
target_percentages = df['Target'].value_counts(normalize=True) * 100

print("Frequency counts for Target:")
display(target_counts)

print("\nPercentage distribution for Target:")
display(target_percentages)

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Target')
plt.title('Distribution of Target Variable')
plt.xlabel('Target (0: No Default, 1: Default)')
plt.ylabel('Frequency')
plt.xticks([0, 1], ['Not Missed next payment', 'Missed next payment'])
plt.show()
```

```python
categorical_features = ['Location', 'EmploymentStatus', 'LoanType']
numerical_features = ['Age', 'Income', 'LoanAmount', 'InterestRate', 'MissedPayments', 'DelaysDays', 'PartialPayments', 'InteractionAttempts', 'SentimentScore', 'ResponseTimeHours', 'AppUsageFrequency', 'WebsiteVisits', 'Complaints']

# Visualize relationship with categorical features
for col in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, hue='Target', palette='viridis')
    plt.title(f'Distribution of Target by {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Target', labels=['Not Default', 'Default'])
    plt.tight_layout()
    plt.show()

# Visualize relationship with numerical features
for col in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='Target', y=col, palette='viridis')
    plt.title(f'Distribution of {col} by Target')
    plt.xlabel('Target (0: Not Default, 1: Default)')
    plt.ylabel(col)
    plt.xticks([0, 1], ['Not Default', 'Default'])
    plt.tight_layout()
    plt.show()
```
